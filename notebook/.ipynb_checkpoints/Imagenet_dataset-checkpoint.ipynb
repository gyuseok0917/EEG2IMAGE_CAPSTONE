```python
def plot_topo(evok_data, title, vlim = (-1, 1)):
    fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 8))
    topo_data = np.mean(evok_data.data, axis = 1)
    topo_data = (topo_data - np.mean(topo_data)) / np.std(topo_data)
    
    im, cn = mne.viz.plot_topomap(
        topo_data, evok_data.info, show = False, cmap = 'coolwarm',
        sensors = False, res = 1000, axes=axes, size = 2, vlim = vlim
    )
    
    axes.set_title(f"{title}")
    cax = fig.add_axes([0.92, 0.33, 0.005, 0.4])
    fig.colorbar(im, cax=cax)
```

```python
plot_topo(lr_eeg, "LR")
plot_topo(hr_eeg, "HR")
```
