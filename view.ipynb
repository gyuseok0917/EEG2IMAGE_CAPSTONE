```python
import re
import matplotlib.pyplot as plt

# 로그 파일 경로
log_path_norm = 'EEGStyleGAN-ADA_CVPR40/out/00023-EEGImageCVPR40-cond-mirror-cifar-bgcfnc/log.txt'

ticks = []
g_losses = []
d_losses = []

# 로그 파일 읽기
with open(log_path_norm, 'r') as file:
    for line in file:
        # 각 라인에서 tick, G_loss, D_loss 값 추출
        match = re.search(r'tick (\d+).*G_loss ([\d\.]+).*D_loss ([\d\.]+)', line)
        if match:
            ticks.append(int(match.group(1)))
            g_losses.append(float(match.group(2)))
            d_losses.append(float(match.group(3)))


print(f'Ticks: {ticks[:5]}')
print(f'G_loss: {g_losses[:5]}')
print(f'D_loss: {d_losses[:5]}')
```

```python
# 데이터 시각화
plt.figure(figsize=(10, 5))

plt.plot(ticks, g_losses, label='G_loss', color='blue')
plt.plot(ticks, d_losses, label='D_loss', color='red')

plt.xlabel('Tick')
plt.ylabel('Loss')
plt.title('G_loss and D_loss over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Softplus 함수 정의
def softplus(x):
    return np.log(1 + np.exp(x))

# 범위 설정
x_values = np.linspace(-10, 10, 100)
y_values = softplus(x_values)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Softplus Function', color='b')
plt.title('Softplus Function')
plt.xlabel('x')
plt.ylabel('Softplus(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 힌지 손실 함수 정의
def hinge_loss(y_true, y_pred):
    return np.maximum(0, 1 + y_true * y_pred)

# 범위 설정
x_values = np.linspace(-10, 10, 100)
y_true = 1  # 예시로 y_true를 1로 설정
y_pred_values = x_values

# 힌지 손실 계산
loss_values = hinge_loss(y_true, y_pred_values)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x_values, loss_values, label='Hinge Loss', color='b')
plt.title('Hinge Loss Function')
plt.xlabel('f(x)')
plt.ylabel('Hinge Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

# 로그 파일 경로
log_file_path = 'EEGStyleGAN-ADA_CVPR40/out/00026-EEGImageCVPR40-cond-mirror-cifar-bgcfnc/log.txt'

# 데이터를 저장할 리스트 초기화
ticks = []
fid_values = []

# 로그 파일 읽기
with open(log_file_path, 'r') as f:
    for line in f:
        # 'results' 줄에서 FID 값 추출
        if '"fid50k_full"' in line:
            match = re.search(r'"fid50k_full": ([\d\.]+)', line)
            if match:
                fid_value = float(match.group(1))
                fid_values.append(fid_value)
        # 'tick' 줄에서 틱 값을 추출
        elif 'tick' in line:
            match = re.search(r'tick (\d+)', line)
            if match:
                tick = int(match.group(1))
                ticks.append(tick)

# 시간 (10틱마다 매트릭 계산)
times = [tick * 10 for tick in ticks if tick % 10 == 0]

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(times, fid_values, marker='o', linestyle='-', color='b')
plt.xlabel('Time (ticks * 10)')
plt.ylabel('FID score')
plt.title('FID score over time')
plt.grid(True)
plt.show()
```

```python
import os

def count_npy_files(directory):
    """
    Given a directory, count the number of .npy files within it.
    
    Args:
    directory (str): The path to the directory to search.
    
    Returns:
    int: The number of .npy files in the directory.
    """
    # Ensure the directory exists
    if not os.path.isdir(directory):
        raise ValueError(f"The directory {directory} does not exist.")
    
    # Count .npy files
    npy_files = [f for f in os.listdir(directory) if f.endswith('.npy')]
    return len(npy_files)

# Example usage
directory_path = 'dataset/new/test'
npy_file_count = count_npy_files(directory_path)
print(f"Number of .npy files in '{directory_path}': {npy_file_count}")
1987+ 7959+ 1994
```

```python
import jsonlines
import matplotlib.pyplot as plt

# JSONL 파일 경로 설정
file_path = 'EEGStyleGAN-ADA_CVPR40/out/00032-EEGImageCVPR40-cond-mirror-cifar-bgcfnc/metric-fid50k_full.jsonl'

# 데이터를 저장할 리스트 초기화
timestamps = []
fid_scores = []

# JSONL 파일 읽기
with jsonlines.open(file_path) as reader:
    for obj in reader:
        fid_scores.append(obj['results']['fid50k_full'])
        timestamps.append(obj['timestamp'])

# 데이터 시각화
plt.figure(figsize=(10, 5))
plt.plot(timestamps, fid_scores, marker='o')
plt.xlabel('Timestamp')
plt.ylabel('FID Score')
plt.title('FID Score Over Time')
plt.grid(True)
plt.xticks(ticks=timestamps[::10], labels=timestamps[::10], rotation=45)  # 10 tick 간격으로 설정
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# JSONL 파일 경로
jsonl_file_path = 'EEGStyleGAN-ADA_CVPR40/out/00032-EEGImageCVPR40-cond-mirror-cifar-bgcfnc/metric-fid50k_full.jsonl'

# 데이터 읽기
timestamps = []
fid_scores = []

with open(jsonl_file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        timestamps.append(data['timestamp'])
        fid_scores.append(data['results']['fid50k_full'])

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(fid_scores) * 10, 10), fid_scores, marker='o', color='blue')
plt.xlabel('Time (ticks * 10)')
plt.ylabel('FID Score')
plt.title('FID Score Over Time')
plt.grid(True)
plt.show()
```
